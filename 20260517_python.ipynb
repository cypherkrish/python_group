```python
text = """Python is a amazing. Python is powerfull. Python is everywhre, Java is also popular, Java runs on billions od devices. Pythn and Java are bot greate languates"
"""
```

```python
print(text)
```

```python
words = text.lower().replace('.', " ").replace(',', " ").replace("-", " ").split()
```

```python
words
```

```python
word_freq = Counter(words)
```

```python
print(word_freq)
```

```python
word_freq.most_common()
```

```python
for i in word_freq.most_common():
    #print(i)
    print(i[0], ":", i[1])
```

```python
for a, b in word_freq.most_common():
        print(a, ":", b)
            
```

```python
# Print only top 5 common word ?

for a, b in word_freq.most_common(5):
        print(f"'{a}' :  {b} times")
```

```python
word_group = ['apple', 'banana', 'apricot', 'blueberry', 'abacodo']
```

```python
word_groups = {}

for word in word_group:
    first_lettter = word[0]
    #print(word, first_lettter)
    if first_lettter not in word_groups:
        word_groups[first_lettter] = []
    word_groups[first_lettter].append(word)
```

```python
word_groups
```

```python
from collections import defaultdict

word_groups_defaultdict = defaultdict(list)

for word in word_group:
    first_lettter = word[0]
    word_groups_defaultdict[first_lettter].append(word)
```

```python
print(dict(word_groups_defaultdict))
```

```python
emp = [
    ('Alice', 'Eng'),
    ('Bob', 'Marketing'),
    ('Charlie', 'Eng'),
    ('Diana', 'HR'),
    ('Eve', 'Marketing'),
    ('Frank', 'Eng')
]
```

```python
# Eng : ['Alice', 'Charlie', 'Frank']
# Marketing: ['Bob', 'Eve']
# HR : ['Diana']
department = {}
for i in emp:
    print(i[1])
    if i[1] not in department:
        department[i[1]] = []
    department[i[1]].append(i[0])
```

```python
department
```

```python
from collections import defaultdict

department = defaultdict(list)
for i in emp:
    print(i[1])
    department[i[1]].append(i[0])
```

```python
department
```

```python
from collections import defaultdict

department = defaultdict(list)
for i, j in emp:
    department[j].append(i)
```

```python
dict(department)
```
