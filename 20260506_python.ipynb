### Enumerate

```python
fruits = ["apple", "banana", "grape", "Orange", "Papaya"]
```

```python
enumerate(fruits)
```

```python
for i in range(len(fruits)):
    print(i," ", fruits[i])
```

```python
for index, value in enumerate(fruits):
    print(index, " ", value)
```

```python
for index, value in enumerate(fruits, start = 1):
    print(index, value)
```

```python
l = [6, 4, 5, 8, 2]

for i , x in enumerate(l):
    print(f"Index: {i},     Value: {x},     List Value: {l[i]}")
```

```python
l = [6, 4, 5, 8, 2]

for i , x in enumerate(l):
    l[i] = x * x
    print(f"Index: {i},     Value: {x},     List Value: {l[i]}")
```
