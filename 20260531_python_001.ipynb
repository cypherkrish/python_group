Passing members of one class to another class

```python
class Emp:
    def __init__(self, id, name, salary):
        self.id = id
        self.name = name
        self.salary = salary

    def display(self):
        print(f"ID: {self.id}, Name: {self.name}, Salary: {self.salary}")
```

```python
class MyClass:

    @staticmethod
    def sal_increment(e):
        e.salary += 1000
        e.display()
```

```python
e = Emp(1, "John Doe", 50000)

MyClass.sal_increment(e)
```

```python
class MyClass:

    @staticmethod
    def sal_increment(e):
        e.salary += 1000
        e.display()
    
    @staticmethod
    def mymethod(x, n):
        result = x ** n
        print(f"{x} raised to the power of {n} is {result}")
```

```python
MyClass.mymethod(2, 3)
MyClass.mymethod(5, 2)
```

Innter classes / Nested classes

```python
class Person:
    def __init__(self):
        self.name = "Alice"  # Name object
        self.db = self.Dob() # DOB object

    def display(self):
        print(f"Name: {self.name}, DB: {self.db}")

    class Dob:
        def __init__(self):
            self.day = 31
            self.month = 5
            self.year = 1985

        def display(self):
            print(f"DOB: {self.day}-{self.month}-{self.year}")
```

```python
p = Person()
p.display()
```

```python
x = p.db
x.display()
```

```python
x = p.db
x.display()
```

```python
x = Person().Dob()
x.display()
print(x.day)
print(x.month)
print(x.year)
```

```python
class Father:
    def __init__(self):
        self.property = 5000000

    def display_property(self):
        print(f"Father's Property: {self.property}")
    

class Son(Father):
    def __init__(self):
        self.property = 2000000

    def display_property(self):
        print(f"Son's Property: {self.property}")
```

```python
s = Son()
s.display_property()
```

```python
class Father:
    def __init__(self, property1=0):
        self.property1 = property1

    def display_property(self):
        print(f"Father's Property: {self.property1}")
    

class Son(Father):
    def __init__(self, property1=0, property2=0):
        super().__init__(property1)
        self.property2 = property2

    def display_property(self):
        print(f"Son's Property: {self.property2 + self.property1}")
```

```python
s = Son(5000000, 2000000)
s.display_property()
```

```python
class Square(object):
    def __init__(self, x):
        self.x = x

    def area(self):
        print(f"Area of the square: {self.x * self.x}")


class Rectangle(Square):
    def __init__(self, x, y):
        super().__init__(x)
        self.y = y

    def area(self):
        super().area()
        print(f"Area of the rectangle: {self.x * self.y}")
```

```python
a = 10
b = 20

r = Rectangle(a, b)
```

```python
r.area()
```

# Single inheritance

```python
class ReserveBank(object):
    cash = 1000000000  # Class variable 

    @classmethod
    def available_cash(cls):
        print(f"Reserve Bank Cash: {cls.cash}")


class StateBank(ReserveBank):
    cash = 500000000  # Class variable

    @classmethod
    def available_cash(cls):
        print(f"State Bank Cash: {cls.cash + ReserveBank.cash}")


class CanaraBank(ReserveBank):
    pass
```

```python
c = CanaraBank()
c.available_cash()
```

```python
s = StateBank()
s.available_cash()
```

# Multiple inheritance

```python
class Father:
    def height(self):
        print("height is 6 feet")

class Mother:
    def color(self):
        print("Color is brown")

class Child(Father, Mother):
    pass
```

```python
c  = Child()
c.height()
c.color()
```

```python
class A(object):
    def __init__(self):
        self.a = 'a'
        print(self.a)

class B(object):
    def __init__(self):
        self.b = 'b'
        print(self.b)

class C(A, B):
    def __init__(self):
        self.c = 'c'
        print(self.c)
        super().__init__()

c = C()
```

```python
class A(object):
    def method(self):
        print("Method in class A")
        super().method()

class B(object):
    def method(self):
        print("Method in class B")
        super().method()

class C(object):
    def method(self):
        print("Method in class C")
        super().method()

class X(A, B):
    def method(self):
        print("Method in class X")
        super().method()

class Y(B, C):
    def method(self):
        print("Method in class Y")
        super().method()

class P(X, Y, C):
    def method(self):
        print("Method in class P")
        super().method()
```

```python
p = P()
p.method()
```
