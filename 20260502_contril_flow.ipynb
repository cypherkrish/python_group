```python
char = input("Enter a character: ")

if char == 'a':
    print("Apple")
elif char == 'b':
    print("Ball")
elif char == "c":
    print("Cat")
elif char == "d":
    print("Dog")
else:
    print("Unknown character")
```

```python
char = input("Enter a character: ")

match char:
    case 'a':
        print("apple")
    case "b":
        print("ball")
    case "c":
        print("cat")
    case "d":
        print("dog")
    case _:
        print("Unknown character")
```

```python
# Print multiples of 7 from 1 to 100
sum = 0
for num in range(1, 101):
    if num % 7 == 0:
        print(f"{num} is the multiple of 7")
        sum = sum + num

print(f"\n{sum}")
```
