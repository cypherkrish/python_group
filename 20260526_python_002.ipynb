```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass  # Placeholder method


class Dog(Animal):
    def speak(self):
        return f"{self.name} says BowBow !!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} says MeowMeow !!"
```

```python
dog = Dog("Puppy")
cat = Cat("Honey")
```

```python
print(dog.speak())
print(cat.speak())
```

```python
class Shape:
    def area(self):
        return "Test" # Placeholder method

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, l, b):
        self.l = l
        self.b = b

    def area(self):
        return self.l * self.b
```

```python
circle = Circle(5)
rect = Rectangle(4, 6)
```

```python
print(circle.area())
print(rect.area())
```

## Encapsulation

```python
# public  => self.name = name
# Accessible from any where

# Protected => self._name = name
# Can not be accessed outside the class directly


# Private => self.__name = name
# Can not be accessed outside the class directly
# To avoid any accessidental access from outside of the class
```

```python
class Person:
    def __init__(self, name, age):
        self._name = name  # protected
        self._age = age # protected
    
    def get_name(self): # getter method
        return self._name

    def set_name(self, name): # Setter method
        if len(name) > 2:
            self._name = name
        else:
            print("Name is not valid")
```

```python
person = Person("abc", 10)
```

```python
person.get_name()
```

```python
person.set_name("xyz")
```

```python
person.get_name()
```
