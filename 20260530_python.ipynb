Classes and Objects : Variables

```python
class Sample:
    def __init__(self):
        self.x = 10 # Public variable => Object Variable

    def modify(self):
        self.x += 1

s1 = Sample()
s2 = Sample()
print('x in s1 = ',s1.x)
print('x in s2 = ', s2.x)

s1.modify()
print('x in s1 = ',s1.x)
print('x in s2 = ', s2.x)
```

```python
class Sample:
    x = 10 # Class variable

    @classmethod
    def modify(cls):
        cls.x += 1


s1 = Sample()
s2 = Sample()
print('x in s1 = ',s1.x)
print('x in s2 = ', s2.x)

s1.modify()
print('x in s1 = ',s1.x)
print('x in s2 = ', s2.x)
```

```python
class Student:
    def __init__(self, n = '', m = 0):
        self.name = n
        self.marks = m
    
    def display(self):
        print("Name: ", self.name)
        print("Marks: ", self.marks)

    def calculate(self):
        if (self.marks >= 600):
            print("First class")
        elif(self.marks >= 500):
            print("Second class")
        elif(self.marks >= 350):
            print ("Third class")
        else:
            print("Failed")
```

```python
n = int(input("How many students: "))

i = 0

while(i < n):
    name = input("Name: ")
    marks = int(input("Marks: "))

    s = Student(name, marks)
    s.display()
    s.calculate()
    i += 1
    print("------------------------")
```

```python
class Student:
   
    # Accessor method
    def getName(self):
        return self.name
    
    # Mutator method
    def setName(self, name):
        self.name = name

        # Accessor method
    def getMarks(self):
        return self.marks
    
    # Mutator method
    def setMarks(self, marks):
        self.marks = marks
```

```python
n = int(input("How many students: "))

i = 0

while(i < n):

    s = Student()

    name = input("Name: ")
    s.setName(name)
    marks = int(input("Marks: "))
    s.setMarks(marks)

    print("Name: ", s.getName())
    print("Marks: ", s.getMarks())
    i += 1
    print("------------------------")
```

```python
# Class method

class Bird:

    wings = 2

    @classmethod  # Decorator
    def fly(cls, name):
        print(f"{name} flies with {cls.wings} wings")
```

```python
Bird.fly("Crow")
Bird.fly("Sparrow")
```

```python
# Static Methods

class MyClass:
    n = 0
    def __init__(self):
        MyClass.n = MyClass.n + 1
    
    @staticmethod
    def noOfObjects():
        print("No .of  objects: ", MyClass.n)
```

```python
import math
class Sample:

    @staticmethod
    def sqare_root(x):
        return math.sqrt(x)
```

```python
import sys
class Bank(object):

    def __init__(self, name, balance=0.0):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("No amount to withdraw")
        else:
            self.balance = self.balance - amount
        return self.balance
    
```

```python
name = input("Name: ")
b = Bank(name)
```

```python
while(True):
    print("d for deposit, w - withdraw, e - Exit")
    choice = input("Your choice: ")
    if choice == 'e' or choice == 'E':
        sys.exit()
    amt = float(input("Amout: "))

    if choice == 'd' or choice == 'D':
        print(f"Total avaialble balance after depositing {amt} is: {b.deposit(amt)}")
    
    elif choice == 'w' or choice =='W':
        print(f"Total avaialble balance after withdrew {amt} is: {b.withdraw(amt)}")
        
```
